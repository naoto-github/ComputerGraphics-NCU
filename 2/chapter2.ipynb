```python
# 画像の読み込み
image = cv2.imread("samples/Lenna.bmp")
```

```python
# BGRからRGBに変換
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# 画像の解像度
print(image.shape)

# 画像を表示
plt.imshow(image_rgb)
```

```python
# 標本化（縦横を1/10に縮小）
scale_factor = 0.1
small_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_NEAREST)

# 画像の解像度
print(small_image.shape)
```

```python
# 標本化した画像を元のサイズにリサイズ
resampled_image = cv2.resize(small_image, image.shape[1::-1], interpolation=cv2.INTER_NEAREST)

# BGRからRGBに変換
resampled_rgb = cv2.cvtColor(resampled_image, cv2.COLOR_BGR2RGB)

# 画像の解像度
print(resampled_rgb.shape)

# 画像を表示
plt.imshow(resampled_rgb)
```

```python
# 画像の読み込み（色深度は8ビット）
image = cv2.imread("samples/Lenna.bmp")

# 画像をグレースケールに変換
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 画像を表示
plt.imshow(image_gray, cmap="gray")
```

```python
# 2ビットに変換
bits = 2
factor = 2 ** (8 - bits)
reduced_image = (image_gray // factor) * factor

# 画像を表示
plt.imshow(reduced_image, cmap="gray")
```

```python
# PGMファイルとして保存する関数
def save_pgm(image, filename):
    height, width = image.shape
    max_val = image.max()

    with open(filename, 'w') as f:
        f.write("P2\n")  # P2はASCII形式を意味する
        f.write(f"{width} {height}\n")  # 画像の幅と高さ
        f.write(f"{max_val}\n")  # 最大のグレースケール値
        for row in image:
            f.write(" ".join(map(str, row)) + "\n")  # ピクセル値を保存
```

```python
# 256x256の画像を生成し、4分割で黒と白を交互に配置する
image = np.zeros((256, 256), dtype=np.uint8)
image[:128, :128] = 0      # 左上
image[:128, 128:] = 255    # 右上
image[128:, :128] = 255    # 左下
image[128:, 128:] = 0      # 右下

# 画像を生成してPGM形式で保存
save_pgm(image, "sample.pgm")
```

```python
# 画像の読み込み
image = cv2.imread("sample.pgm")

# 画像をグレースケールに変換
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 画像を表示
plt.imshow(image_gray, cmap="gray")
```
