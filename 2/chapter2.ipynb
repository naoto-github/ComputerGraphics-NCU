```python
# PGM
def save_pgm(image, filename):
    height, width = image.shape
    max_val = image.max()

    with open(filename, 'w') as f:
        f.write("P2\n")  # P2ASCII
        f.write(f"{width} {height}\n")  # 
        f.write(f"{max_val}\n")  # 
        for row in image:
            f.write(" ".join(map(str, row)) + "\n")  # 
```

```python
# 256x2564
image = np.zeros((256, 256), dtype=np.uint8)
image[:128, :128] = 0      # 
image[:128, 128:] = 255    # 
image[128:, :128] = 255    # 
image[128:, 128:] = 0      # 

# PGM
save_pgm(image, "sample.pgm")
```
