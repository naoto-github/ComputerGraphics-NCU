# 「Mandrill.bmp」を読み込み，画像の解像度を1/4に縮小し，その後元のサイズに戻すプログラムを作成せよ．

```python
# 画像の読み込み
image = cv2.imread("samples/Mandrill.bmp")

# 標本化（縦横を1/10に縮小）
scale_factor = 0.25
small_image = cv2.resize(image, None, fx=scale_factor, fy=scale_factor, interpolation=cv2.INTER_NEAREST)

# 画像の解像度
print(small_image.shape) # (26, 26, 3)

# 標本化した画像を元のサイズにリサイズ
resampled_image = cv2.resize(small_image, image.shape[1::-1], interpolation=cv2.INTER_NEAREST)

# BGRからRGBに変換
resampled_rgb = cv2.cvtColor(resampled_image, cv2.COLOR_BGR2RGB)

# 画像の解像度
print(resampled_rgb.shape) # (256, 256, 3)

# 画像を表示
plt.imshow(resampled_rgb)
```

# 「Mandrill.bmp」を読み込み，画像を色深度2ビットのグレースケールに変換して表示するプログラムを作成せよ．

```python
# 画像の読み込み（色深度は8ビット）
image = cv2.imread("samples/Mandrill.bmp")

# 画像をグレースケールに変換
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2ビットに変換
bits = 2
factor = 2 ** (8 - bits)
reduced_image = (image_gray // factor) * factor

# 画像を表示
plt.imshow(reduced_image, cmap="gray")
```

# 「Mandrill.bmp」を読み込み，JPEG形式で品質10と品質90で保存後，それぞれの画像ファイルサイズを比較するプログラムを作成せよ．

```python
import os

# BMP画像の読み込み
bmp_image = cv2.imread("samples/Mandrill.bmp")

# JPEG画像として保存（品質10）
cv2.imwrite("Mandrill_quality_10.jpg", bmp_image, [cv2.IMWRITE_JPEG_QUALITY, 10])

# JPEG画像として保存（品質10）
cv2.imwrite("Mandrill_quality_90.jpg", bmp_image, [cv2.IMWRITE_JPEG_QUALITY, 90])

# 保存したJPEG画像の読み込み
jpeg_image_10 = cv2.imread("Mandrill_quality_10.jpg")

# RGBに変換（OpenCVはBGR形式のため）
jpeg_image_10_rgb = cv2.cvtColor(jpeg_image_10, cv2.COLOR_BGR2RGB)

# 保存したJPEG画像の読み込み
jpeg_image_90 = cv2.imread("Mandrill_quality_90.jpg")

# RGBに変換（OpenCVはBGR形式のため）
jpeg_image_90_rgb = cv2.cvtColor(jpeg_image_90, cv2.COLOR_BGR2RGB)

# 元画像と圧縮画像を表示
fig, axes = plt.subplots(1, 2, tight_layout=True)
axes[0].imshow(jpeg_image_10_rgb)
axes[1].imshow(jpeg_image_90_rgb)

print(f"{os.path.getsize("Mandrill_quality_10.jpg")} bytes")
print(f"{os.path.getsize("Mandrill_quality_90.jpg")} bytes")
```

# Mandrill.bmp」を読み込み，PGM形式（ASCII）で保存するプログラムを作成せよ．

```python
# PGMファイルとして保存する関数
def save_pgm(image, filename):
    height, width = image.shape
    max_val = image.max()

    with open(filename, 'w') as f:
        f.write("P2\n")  # P2はASCII形式を意味する
        f.write(f"{width} {height}\n")  # 画像の幅と高さ
        f.write(f"{max_val}\n")  # 最大のグレースケール値
        for row in image:
            f.write(" ".join(map(str, row)) + "\n")  # ピクセル値を保存
```

```python
# 画像の読み込み
image = cv2.imread("samples/Mandrill.bmp")

# 画像をBGRからRGBに変換
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# PGM形式で保存
save_pgm(image_rgb, "Mandrill.pgm")
```
