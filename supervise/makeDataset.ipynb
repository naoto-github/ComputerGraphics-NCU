```python
import pandas as pd
import numpy as np
import os
import shutil
import zipfile
```

```python
df = pd.read_csv("df.csv")
df = df.drop(columns="Unnamed: 0")
display(df)
```

```python
size = 11
datasets = []

for seed in range(size):
    sample_df = df.sample(n=100, replace=False, random_state=seed, ignore_index=False)
    sample_df.reset_index(drop=True, inplace=True)
    display(sample_df)
    datasets.append(sample_df)
```

```python
# フォルダの作成
for i in range(size):
    # メインフォルダの名前
    main_folder = f"samples/sample_{i}"
    kadai_folder = f"kadai/sample_{i}"
    
    main_sub_folders = ["images", "masks", "collages"]
    kadai_sub_folders = ["images"]
    
    if os.path.exists(main_folder):
        shutil.rmtree(main_folder)

    if os.path.exists(kadai_folder):
        shutil.rmtree(kadai_folder)
    
    os.makedirs(main_folder)
    os.makedirs(kadai_folder)
    
    for folder in main_sub_folders:
        os.makedirs(os.path.join(main_folder, folder))

    for folder in kadai_sub_folders:
        os.makedirs(os.path.join(kadai_folder, folder))
```

```python
# ファイルのコピー
for i in range(size):

    target_folder = f"samples/sample_{i}/"
    kadai_folder = f"kadai/sample_{i}/"
    target_df = datasets[i]

    for index,row in target_df.iterrows():
        image_path = row["images"]
        mask_path = row["masks"]
        collage_path = row["collages"]
        filename = str(index).zfill(3) + ".png"
        shutil.copy("dataset/" + image_path, target_folder + "images/" + filename)
        shutil.copy("dataset/" + mask_path, target_folder + "masks/" + filename)
        shutil.copy("dataset/" + collage_path, target_folder + "collages/" + filename)
        shutil.copy("dataset/" + image_path, kadai_folder + "images/" + filename)
```

```python
def zip_folder(folder_path, output_path):
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # フォルダ内の全ファイルとサブフォルダを走査
        for root, dirs, files in os.walk(folder_path):
            for file in files:
                # ファイルのフルパスを取得
                file_path = os.path.join(root, file)
                # ZIPファイル内のパスを設定（相対パス）
                arcname = os.path.relpath(file_path, os.path.dirname(folder_path))
                # ファイルをZIPに追加
                zipf.write(file_path, arcname)
```

```python
for i in range(size):
    folder_path = f"kadai/sample_{i}/"
    output_path = f"kadai/sample_{i}.zip"
    zip_folder(folder_path, output_path)
```
